```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
import math
# Load pylab
%pylab inline

destinations = pd.read_csv("data/my_submission.csv")
test = pd.read_csv("data/test.csv")
train = pd.read_csv("data/train.csv")
firstColumnInd = destinations["Id"]
```

```python
train["dteday"] = pd.to_datetime(train["dteday"])
#train["year"] = train["dteday"].dt.year
#train["day"] = train["dteday"].dt.day

train.drop('dteday', axis=1, inplace=True)
#train.drop('yr', axis=1, inplace=True)
train["cnt"] = log(train["cnt"])

train["cmnth"] = cos(train["mnth"]/(2*math.pi))
train["smnth"] = sin(train["mnth"]/(2*math.pi))
train["chr"] = cos(train["hr"]/(2*math.pi))
train["shr"] = sin(train["hr"]/(2*math.pi))
train["cseason"] = cos(train["season"]/(2*math.pi))
train["sseason"] = sin(train["season"]/(2*math.pi))
train["cweekday"] = cos(train["weekday"]/(2*math.pi))
train["sweekday"] = sin(train["weekday"]/(2*math.pi))


train.drop('hr', axis=1, inplace=True)
train.drop('mnth', axis=1, inplace=True)


test["dteday"] = pd.to_datetime(test["dteday"])
#test["year"] = test["dteday"].dt.year

test["cmnth"] = cos(test["mnth"]/(2*math.pi))
test["smnth"] = sin(test["mnth"]/(2*math.pi))
test["chr"] = cos(test["hr"]/(2*math.pi))
test["shr"] = sin(test["hr"]/(2*math.pi))
test["cseason"] = cos(test["season"]/(2*math.pi))
test["sseason"] = sin(test["season"]/(2*math.pi))
test["cweekday"] = cos(test["weekday"]/(2*math.pi))
test["sweekday"] = sin(test["weekday"]/(2*math.pi))


test.drop('hr', axis=1, inplace=True)
test.drop('mnth', axis=1, inplace=True)



#test["day"] = test["dteday"].dt.day
#test.drop('yr', axis=1, inplace=True)
test.drop('dteday', axis=1, inplace=True)

```

```python
train["hum"] = preprocessing.scale(train["hum"])
train["temp"] = preprocessing.scale(train["temp"])
train["atemp"] = preprocessing.scale(train["atemp"])
train["windspeed"] = preprocessing.scale(train["windspeed"])


test["hum"] = preprocessing.scale(test["hum"])
test["temp"] = preprocessing.scale(test["temp"])
test["atemp"] = preprocessing.scale(test["atemp"])
test["windspeed"] = preprocessing.scale(test["windspeed"])

```

```python
X = train[['cseason','sseason','yr','cmnth','chr','smnth','shr','cweekday','sweekday','holiday','workingday','weathersit','temp','atemp','hum','windspeed']]
y= train['cnt']

X_test_set =test[['cseason','sseason','yr','cmnth','chr','smnth','shr','cweekday','sweekday','holiday','workingday','weathersit','temp','atemp','hum','windspeed']]
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=5)
print X_train.shape
print X_test.shape

```

```python
from sklearn.neighbors import KNeighborsRegressor

kneighbor_regression = KNeighborsRegressor(n_neighbors=6)
kneighbor_regression.fit(X_train, y_train)


print kneighbor_regression.score(X_train, y_train)
print kneighbor_regression.score(X_test, y_test)
```

```python
y_pred_test_real = kneighbor_regression.predict(X_test_set)

y_pred_test_real= exp(y_pred_test_real)
print y_pred_test_real
```
